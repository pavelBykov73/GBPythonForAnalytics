## Задание 3. 
**Условие.**

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
plt.ylabel('Стоимость')
plt.xlabel('Жилая площадь');
```

Есть достаточно линейная зависимость стоимости от жилой площади. В основном объекты сосредоточены в райное 2000sqft. При этом причина разброса стоимости при одинаковой площади должна быть дополнительно изучена.

```python
sns.jointplot(x=df['yr_built'], y=df['price'], kind='kde');
plt.ylabel('Стоимость')
plt.xlabel('Год постройки');
```

Стоимость тем больше, чем новее дом. Но на рынке также котируются и сравнительно старые дома 1960 годов постройки

```python
sns.jointplot(x=df['condition'], y=df['price'], kind='reg');
plt.ylabel('Стоимость')
plt.xlabel('Состояние');
```

Плохое состояние дома снижает его стоимость. Чем лучше состояние дома, больше стоимость и ее разброс.

```python
sns.jointplot(x=df['lat'], y=df['price'], kind='kde');
plt.ylabel('Price')
plt.xlabel('Lattitude');
```

Стоимость домов зависит от расположения. Есть несколько локальных максимумов По видимому это связано с агломерациями (города, поселки)

```python
properties = df[['price', 'sqft_living', 'floors', 'grade', 'condition',
           'waterfront', 'view', 'yr_built', 'sqft_lot', 'bedrooms',
           'bathrooms', 'long']]
```

```python
corr_matrix = properties.corr()
```

```python
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, linewidths = 0.5, cmap='coolwarm', fmt="1.2f");
```

Рассматривая тепловую карту кросс таблицы факторов (читаем верхнюю строку - влияние на цену), можно заметить, что наибольшее влияние оказывает жилая площадь и grade (оценка). Также влияет количество спален и ванных комнат (это связано с площадью (0.75)), вид (View). Местоположение практически не влияет.
